```python
import matplotlib.pyplot as plt

python_versions = [
    ("0.9.0",1991),
    ("1.0",1994),
    ("2.0",2000),
    ("2.6",2008),
    ("3.0",2008),
    ("3.1",2009),
    ("2.7",2010),
    ("3.2",2011),
    ("3.3",2012),
    ("3.4",2014),
    ("3.5",2015),
    ("3.6",2016),
    ("3.7",2018),
    ("3.10",2020),
    ("3.11",2021),
    ("3.12",2022),
    ("3.13",2023),
    ("3.14",2024),
    ]

# Extract data for plotting
versions = [v[0] for v in python_versions]
years = [v[1] for v in python_versions]
y_versions = [int(v[0][0]) for v in python_versions]

# colors based on versions
version_colors = {
    0: 'red',
    1: 'blue',
    2: 'green',
    3: 'pink'
}
colors = [version_colors[major_version] for major_version in y_versions]

# Plot
plt.figure(figsize=(14,6))
plt.scatter(years, y_versions, color=colors, zorder=5)

# Label points
y_offset = .2
for y, (version, year) in zip(y_versions, python_versions):
    plt.text(year,
             y+y_offset,
             version,
             fontsize=10,
             verticalalignment='top',
             horizontalalignment='center')

# Customize plot
plt.title("Python Versions Release Over Time", size=15)
plt.xticks(years, rotation=90)
plt.ylim(0, 4)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.yticks([])  # Remove y-ticks
#plt.gca().get_yaxis().set_ticks([])  # Remove y-tick label


# Show grid on x-axis
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```

### Data-driven finance

Advances in technology vs. financial business practices
* access to all finance data available
* real-time or near real-time
* before:
  - what traders and portfolio managers could learn in conversations / read in papers
  - traditional finance based on daily close (appr **7500** datapoints/year)
  - structured regular planning decision making
  - overnight batch runs in the backoffice
* now:
  - machines deal with volume and velocity of financial info
  - smaller time scale (millions of datapoints, ex. **15000** datapoints/hour for APPL)
  - chalenges:
    - dataprocessing (24/7 for some instruments and a lot going on during the day)
    - analytics speed (decisions in milliseconds => real time analytics)
    - some activities have moved to real time front-office (ex. some risk mngt)
    - theoretical foundations (traditional finance does not work in this scale)
  - algorithms explore **market microstructure elements**:
    - order flow
    - bid-ask spreads

* future:
  - **AI-first finance**: flexible, parameterizable ML and DL algorithms replace traditional financial theory
